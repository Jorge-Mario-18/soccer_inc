```python
#Import packages
import pandas as pd
```

```python
#Read CSV
logins = pd.read_csv('LoginHistory.csv')
```

```python
#Drop Columns
logins.drop(columns = ['Logged out', 'Location', 'Client Type'])
```

```python
# Retrives login data from 'Logged in' column 
loginTime = logins['Logged in']
```

```python
#Seperates the the 'Date' and 'Time' of login
time = []
date = []

for i in loginTime:
  txt = i.split(' ', 1)
  time.append(txt[1])
  date.append(txt[0])
```

```python
# Creates the number of login occurances by date
rep = {}
for i in date:
  rep[i] = date.count(i)
```

```python
# Creates data frame with Columns 
df = pd.DataFrame(columns=['Day', 'Count'])
```

```python
# Seperates the values from the list into x and y values 
lists = sorted(rep.items())
x, y = zip(*lists)
```

```python
# Adds the data for 'Day' and 'Count' columns
df['Day'] = x
df['Count'] = y
```

```python
# Creates visual graph of the data
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.ylabel('Login Count')
plt.xlabel('Day of Login Occurance')
plt.savefig('DailyLogins.png')
plt.show()
```

```python
# Drops outliers of more then 80 logins per day
df.drop(df[df['Count'] > 80].index, inplace = True)
df = df.reset_index(drop=True)
```

```python
# New plot with outliers removed past 80 logins 
import matplotlib.pyplot as plt

plt.plot(df['Day'], df['Count'])
plt.ylabel('Login Count')
plt.xlabel('Day of Login Occurance')
plt.savefig('DailyLogins.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ax = chart20.plot(x='Quater', y='Average', ylim=(0, 50), label='2020')

ax = chart21.plot(ax=ax, x='Quater', y='Average', ylim=(0, 50), label='2021')

chart22.plot(ax=ax, x='Quater', y='Average', ylim=(0, 50), label='2022', xlabel='Average for each Quarter')
plt.grid()
plt.ylabel('Login Count')
plt.xticks(np.arange(0, 1, step=0.1))
plt.xticks(np.arange(4))
plt.xticks([0, 1, 2, 3], ['Q1', 'Q2', 'Q3', 'Q4'],
        rotation=20)
plt.xticks()
plt.savefig('Quater-Average.png')
```
